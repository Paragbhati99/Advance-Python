```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

'''Lab1: Analyze the relationship between the size of houses (measured in square footage) 
and their selling prices in a particular neighborhood. 
You have collected data on various houses in that neighborhood.
Create a scatter plot using the below data and share your conclusion/analysis.'''

#Input:
squarefootage = np.array([1200, 1400, 1600, 1800, 2000, 2200, 2400, 2600, 2800, 3000])
selling_prices = np.array([250, 290, 315, 380, 410, 450, 500, 525, 570, 610])

# Plotting the data
plt.figure(figsize=(10, 6))
plt.scatter(squarefootage, selling_prices, color='blue', label='Data Points')

# Adding labels and title
plt.title('Relationship Between House Size and Selling Price', fontsize=16)
plt.xlabel('Square Footage (sq ft)', fontsize=12)
plt.ylabel('Selling Price ($K)', fontsize=12)

# Adding a trend line (linear regression)
m, b = np.polyfit(squarefootage, selling_prices, 1)  # Slope and intercept
plt.plot(squarefootage, m * squarefootage + b, color='red', label=f'Trend Line: y = {m:.2f}x + {b:.2f}')

plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
'''Lab2: Create a pie chart to visualize the distribution of your monthly income by source. 
You have collected data on the various sources of your income, such as salary, freelance work, investments, 
and rental income. Share your conclusion/analysis.'''

#Input:
income_sources = ['Salary', 'Freelance', 'Investments', "Rental", "Other"] 
monthly_income=[5000, 1500, 1000, 600, 400]

# Plotting the pie chart
plt.figure(figsize=(8, 8))
colors = ['skyblue', 'orange', 'lightgreen', 'gold', 'pink']
explode = [0.1, 0, 0, 0, 0]  # Highlight the largest source (Salary)

plt.pie(monthly_income, labels=income_sources, autopct='%1.1f%%', startangle=140,
        colors=colors, explode=explode, shadow=True, textprops={'fontsize': 12})

# Adding title
plt.title('Monthly Income Distribution by Source', fontsize=16)
plt.tight_layout()
plt.show()
```
