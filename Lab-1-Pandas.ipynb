```python
import pandas as pd
import matplotlib.pyplot as plt
#1.Visualize the daily temperature changes over time in a city and give your conclusion.

# Input data
days = list(range(1, 32))
temperature = [65, 68, 70, 72, 75, 76, 78, 80, 81, 79, 75, 72, 70, 68, 67, 69, 70, 73, 75, 76, 78, 80, 81, 82, 83, 82, 80, 78, 76, 74, 71]

# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(days, temperature, marker='o', linestyle='-', color='b', label='Temperature (°F)')

# Adding labels and title
plt.title('Daily Temperature Changes Over a Month', fontsize=16)
plt.xlabel('Days of the Month', fontsize=12)
plt.ylabel('Temperature (°F)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.xticks(days, fontsize=10)
plt.yticks(fontsize=10)
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# 2. Create a line plot to Visualize the daily closing price of a stock over a year and give your conclusion.

import matplotlib.pyplot as plt

# Input data
days = list(range(1, 78))
stock_prices = [
    100, 105, 110, 115, 112, 120, 118, 125, 128, 130, 132, 135,
    138, 140, 142, 144, 145, 148, 150, 155, 160, 158, 162, 165,
    170, 172, 175, 178, 180, 182, 185, 188, 190, 192, 195, 198,
    200, 198, 195, 193, 190, 188, 185, 182, 180, 178, 175, 172,
    170, 168, 165, 162, 160, 158, 155, 152, 150, 148, 145, 143,
    140, 138, 135, 132, 130, 128, 125, 123, 120, 118, 115, 112,
    110, 108, 105, 103, 100
]

# Plotting the data
plt.figure(figsize=(12, 6))
plt.plot(days, stock_prices, marker='o', linestyle='-', color='green', label='Closing Price')

# Adding labels and title
plt.title('Daily Closing Price of Stock Over Time', fontsize=16)
plt.xlabel('Days', fontsize=12)
plt.ylabel('Stock Price ($)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.xticks(range(0, len(days)+1, 5), fontsize=10)
plt.yticks(fontsize=10)
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
